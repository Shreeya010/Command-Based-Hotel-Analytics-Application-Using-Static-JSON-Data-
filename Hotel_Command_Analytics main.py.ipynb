```python
# Expand the 'data' column
df_data = pd.json_normalize(df['data'])

df_data
```

```python
df_sales = df_data.explode('sales')
df_sales
```

```python
sales_details = pd.json_normalize(df_sales['sales'])
sales_details
```

```python
dataframe = pd.concat(
    [df_sales[['date']].reset_index(drop=True),
     sales_details.reset_index(drop=True)],
    axis=1
)

dataframe
```

```python
dataframe.info()
```

```python
def intent_detection(command):
    command = command.lower()
    if "today" in command:
        return "TODAY_SALE"
    elif "payment" in command:
        return "PAYMENT"
    elif "cancel" in command:
        return "CANCELLED"
    elif "predict" in command:
        return "PREDICT"
    else:
        return "EXIT"
```

```python
print(intent_detection("What is today's sale?"))
print(intent_detection("Show payment wise sale"))
print(intent_detection("How many bills were cancelled"))
print(intent_detection("Predict cancellation risk"))
```

```python
from datetime import date

today = str(date.today())

today_sale = dataframe[(dataframe['date'] == today) & (dataframe['is_cancelled'] == False)]['amount'].sum()

print("Today's Sales", today_sale)
```

```python
payment_wise_sale = (dataframe[dataframe['is_cancelled'] == False].groupby('payment_mode')['amount'].sum())

payment_wise_sale
```

```python
cancelled_bills = dataframe['is_cancelled'].sum()
print("Cancelled Bills:", cancelled_bills)
```

```python
from sklearn.preprocessing import LabelEncoder

le_payment = LabelEncoder()
le_room = LabelEncoder()

dataframe['payment_encoded'] = le_payment.fit_transform(dataframe['payment_mode'])
dataframe['room_encoded'] = le_room.fit_transform(dataframe['room_type'])
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
x= dataframe[['amount', 'payment_encoded', 'room_encoded']]
y= dataframe['is_cancelled']

model = LogisticRegression()
model.fit(x,y)
```

```python
sample = [[5000, 1, 1]]
cancellation_risk = model.predict_proba(sample)[0][1]

print(f"Cancellation Risk: {cancellation_risk*100:.2f}%")
```
